## Parte 2: Obtener los datos deseados

* Buscar dentro del contenido HTML y extraer la información.

```python
# Extracting the description
description = soup.find("meta", {"name": "description"})["content"]

# Extracting the ingredients
ingredients_section = soup.find_all("li", class_="mm-recipes-structured-ingredients__list-item")
ingredients = [ingredient.get_text().strip() for ingredient in ingredients_section]

# Extracting the instructions
instructions_section = soup.find_all("p", class_="comp mntl-sc-block mntl-sc-block-html")
instructions = [instruction.get_text().strip() for instruction in instructions_section]

# Extracting the nutrition information
nutrition_section = soup.find_all("span", class_="mm-recipes-nutrition-facts-label__nutrient-name mm-recipes-nutrition-facts-label__nutrient-name--has-postfix")
nutrition_facts = [fact.parent.get_text().strip().replace('\n', ' ') for fact in nutrition_section]

# Print the extracted information
print("Recipe Title:", title)
print("Description:", description)
print("Ingredients:")
for ingredient in ingredients:
    print("-", ingredient)
print("Instructions:")
for i, instruction in enumerate(instructions, 1):
    print(f"{i}. {instruction}")
print("Nutrition Facts:")
for fact in nutrition_facts:
    print("-", fact)
```

## Parte 3: Obtener enlaces relacionados
* Encontrar links a otras recetas para completar el corpus

```python
# Find all the links to other recipes
recipe_links = soup.find_all("a", href=True)

# Filter and print only the links that are likely to be recipes
recipe_urls = []
for link in recipe_links:
    href = link['href']
    if "recipe" in href:
        recipe_urls.append(href)

# Print the recipe URLs
print("Linked Recipes:")
for url in recipe_urls:
    print(url)
```

## Parte 4: Hacer RAG con las recetas obtenidas
* Una vez que se ha construido el corpus, implementar y desplegar RAG para realizar búsquedas en el corpus

```python
import re

recipe_urls_clean = set()
for link in recipe_links:
    href = link['href']
    if re.match(r'https://www\.allrecipes\.com/recipe/\d+/', href):
        recipe_urls_clean.add(href)

recipe_urls_clean = list(recipe_urls_clean)
print(f"Recetas: {len(recipe_urls_clean)}")
for url in recipe_urls_clean:
    print(url)
```

```python
import os
import re
import time
import requests
from urllib.parse import urlparse

output_dir = "data_recipes"
os.makedirs(output_dir, exist_ok=True)

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 "
                  "(KHTML, like Gecko) Chrome/126.0 Safari/537.36"
}

downloaded_files = []

for idx, url in enumerate(recipe_urls_clean, start=1):
    try:
        response = requests.get(url, headers=headers, timeout=20)
        response.raise_for_status()

        path = urlparse(url).path.strip("/")
        slug = path.split("/")[-1] if path else f"recipe_{idx}"
        recipe_id = path.split("/")[-2] if len(path.split("/")) >= 2 else str(idx)
        filename = re.sub(r"[^a-zA-Z0-9_-]", "_", f"{recipe_id}_{slug}") + ".html"
        filepath = os.path.join(output_dir, filename)

        with open(filepath, "w", encoding="utf-8") as f:
            f.write(response.text)

        downloaded_files.append({
            "url": url,
            "file": filepath,
            "status": "ok"
        })

        print(f"Descargado: {url} -> {filepath}")
        time.sleep(1)

    except Exception as e:
        downloaded_files.append({
            "url": url,
            "file": None,
            "status": f"error: {e}"
        })
        print(f"Error descargando {url}: {e}")

print(f"\nHTML descargados correctamente: {sum(1 for x in downloaded_files if x['status'] == 'ok')}")
print(f"Errores: {sum(1 for x in downloaded_files if x['status'] != 'ok')}")
```

```python
def extraer_receta(soup):
    if not isinstance(soup, BeautifulSoup):
        soup = BeautifulSoup(soup, "html.parser")

    titulo_tag = soup.find("meta", {"property": "og:title"})
    titulo = titulo_tag.get("content") if titulo_tag and titulo_tag.get("content") else "Sin titulo"

    desc_tag = soup.find("meta", {"name": "description"})
    desc = desc_tag.get("content") if desc_tag and desc_tag.get("content") else ""

    ings = soup.find_all("li", class_="mm-recipes-structured-ingredients__list-item")
    ingredientes = [ing.get_text().strip() for ing in ings] if ings else []

    pasos = soup.find_all("p", class_="comp mntl-sc-block mntl-sc-block-html")
    instrucciones = [p.get_text().strip() for p in pasos] if pasos else []

    nut_facts = soup.find_all("span", class_="mm-recipes-nutrition-facts-label__nutrient-name mm-recipes-nutrition-facts-label__nutrient-name--has-postfix")
    nutrition_facts = [fact.parent.get_text().strip().replace('\n', ' ') for fact in nut_facts] if nut_facts else []

    return {
        "title": titulo,
        "description": desc,
        "ingredients": ingredientes,
        "instructions": instrucciones,
        "nutrition_facts": nutrition_facts
    }
```

```python
import glob
from bs4 import BeautifulSoup

corpus = []

downloaded_files = glob.glob("data_recipes/*.html")

print(f"Archivos encontrados: {len(downloaded_files)}")

for filepath in downloaded_files:
    with open(filepath, "r", encoding="utf-8") as file:
        html_content = file.read()

    recipe = extraer_receta(html_content)
    recipe["file"] = filepath
    corpus.append(recipe)

print(f"Recetas cargadas en corpus: {len(corpus)}")
```

```python
documents = []

for recipe in corpus:
    text = f"""
    Title: {recipe['title']}
    Description: {recipe['description']}
    Ingredients: {', '.join(recipe['ingredients'])}
    Instructions: {' '.join(recipe['instructions'])}
    Nutrition facts: {', '.join(recipe['nutrition_facts'])}
    """
    documents.append(text.strip())
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
embeddings = model.encode(documents, batch_size=120, show_progress_bar=True)

print(f"Shape: {embeddings.shape}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

def search_model(query, model, embeddings, df, k=5):
    query_embedding = model.encode([query])

    similarities = cosine_similarity(query_embedding, embeddings)[0]
    top_k_idx = similarities.argsort()[::-1][:k]

    results = []

    for rank, idx in enumerate(top_k_idx, start=1):
        results.append({
            "rank": rank,
            "title": df.iloc[idx]["title"],
            "description": df.iloc[idx]["description"],
            "ingredients": df.iloc[idx]["ingredients"],
            "instructions": df.iloc[idx]["instructions"],
            "nutrition_facts": df.iloc[idx]["nutrition_facts"]
        })

    results = pd.DataFrame(results)

    return results
```

```python
query = "Give me recipes with chicken and rosemary"
```

```python
results = search_model(query, model, embeddings, pd.DataFrame(corpus), k=10)
results
```

```python
prompt = (
    f"Eres un chatbot coversacional. Mi usuario quiere saber sobre {query}, y mi sistema de recuperacion de la informacion encontro esto {results}."
    f"Dame la respuesta en base al contexto dado por el sistema de recuperacion de la informacion, pero no le hagas saber al usuario sobre el sistema de recuperacion, solo respondele la pregunta")
```
