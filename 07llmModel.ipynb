## Integración llm con sistema de recuperación de la información de embeddings

```python
import pandas as pd
import kagglehub
data_path = kagglehub.dataset_download("stefanoleone992/rotten-tomatoes-movies-and-critic-reviews-dataset")

movies_path = data_path + '\\' + 'rotten_tomatoes_movies.csv'
movies = pd.read_csv(movies_path)
reviews_path = data_path + '\\' + 'rotten_tomatoes_critic_reviews.csv'
reviews = pd.read_csv(reviews_path)
```

```python
print(f"Movies: {movies.shape}")
print(f"Reviews: {reviews.shape}")
```

```python
movies.head(2)
```

```python
reviews.head(2)
```

```python
df = movies.merge(
    reviews[[
        "rotten_tomatoes_link",
        "critic_name",
        "publisher_name",
        "review_type",
        "review_score",
        "review_content"
    ]],
    on="rotten_tomatoes_link",
    how="left"
)
```

```python
df.head(5)
```

```python
df["document"] = (
    df["movie_title"].fillna("") + " " +
    df["movie_info"].fillna("") + " " +
    df["content_rating"].fillna("") + " " +
    df["genres"].fillna("") + " " +
    df["directors"].fillna("") + " " +
    df["authors"].fillna("") + " " +
    df["actors"].fillna("") + " " +
    df["original_release_date"].fillna("") + " " +
    df["runtime"].fillna("").astype(str) + " " +
    df["production_company"].fillna("")
).str.strip()
```

```python
df = df[df["document"] != ""].reset_index(drop=True)
print(len(df))
df["document"].head(5)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download("wordnet", quiet=True)
nltk.download("omw-1.4", quiet=True)
```

```python
import re

stop_words = set(stopwords.words("english"))
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    """
    Preprocesamiento básico para el corpus.
    - Conversión a minúsculas
    - Eliminación de puntuación
    - Eliminación de espacios redundantes
    """
    if not isinstance(text, str):
        return ""
    text = text.lower()                          # minúsculas
    text = re.sub(r"[^\w\s]", "", text)          # eliminar puntuación
    text = re.sub(r"\s+", " ", text).strip()     # espacios redundantes

    words = text.split()
    words = [w for w in words if w not in stop_words]

    # 6. Lematización
    words = [lemmatizer.lemmatize(w) for w in words]

    # 7. Unir de vuelta a string
    return " ".join(words)
```

```python
df = df.sample(frac= 0.1, random_state=42).reset_index(drop=True)

# Aplicar preprocesamiento a la columna document
df["document"] = df["document"].apply(preprocess)

print(f"Documentos para indexar: {len(df)}")
print(f"\nEjemplo preprocesado:\n{df['document'].iloc[0:10]}")
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

# 1. Cargar modelo preentrenado
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# 2. Lista de documentos preprocesados
documents = df["document"].tolist()

# 3. Generar embeddings
print("Calculando embeddings...")
embeddings = model.encode(documents, batch_size=120, show_progress_bar=True)
embeddings = np.array(embeddings)

print(f"Shape de embeddings: {embeddings.shape}")

# 4. Guardar en disco para no recalcular
np.save("embeddings_minilm.npy", embeddings)
print("Embeddings guardados en embeddings_minilm.npy")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def search_model(query, model, embeddings, df, k=5):
    query_processed = preprocess(query)
    query_embedding = model.encode([query_processed])

    similarities = cosine_similarity(query_embedding, embeddings)[0]
    top_k_idx = similarities.argsort()[::-1][:k]

    results = []

    for rank, idx in enumerate(top_k_idx, start=1):
        results.append({
            "Ranking": rank,
            "ID": df["rotten_tomatoes_link"].iloc[idx],
            "Título": df["movie_title"].iloc[idx],
            "Información": df["movie_info"].iloc[idx],
            "Clasificación": df["content_rating"].iloc[idx],
            "Géneros": df["genres"].iloc[idx],
            "Directores": df["directors"].iloc[idx],
            "Autores": df["authors"].iloc[idx],
            "Actores": df["actors"].iloc[idx],
            "Fecha estreno": df["original_release_date"].iloc[idx],
            "Duración": df["runtime"].iloc[idx],
            "Productora": df["production_company"].iloc[idx],
            "Similitud": round(float(similarities[idx]), 4)
        })

    return pd.DataFrame(results)
```

```python
query = "science fiction movie with advanced technology"
```

```python
embeddings_a = embeddings
results = search_model(query, model, embeddings_a, df, k=100)
display(results)
```

```python
prompt = (
    f"Eres un chatbot coversacional. Mi usuario quiere saber sobre {query}, y mi sistema de recuperacion de la informacion encontro esto {results}."
    f"Dame la respuesta en base al contexto dado por el sistema de recuperacion de la informacion")
```
