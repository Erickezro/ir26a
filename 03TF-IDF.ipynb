## Comprensión de TF-IDF

TF-IDF (Frecuencia de Término–Frecuencia Inversa de Documento) es un método estadístico utilizado en el procesamiento del lenguaje natural y la recuperación de información para evaluar la importancia de una palabra en un documento en relación con una colección más amplia de documentos. TF-IDF combina dos componentes:

**1. Frecuencia de término (FT):** Mide con qué frecuencia aparece una palabra en un documento. Una mayor frecuencia indica mayor importancia. Si un término aparece con frecuencia en un documento, es probable que sea relevante para su contenido.

$$TF(t, d) = \frac{\text{Number of times term } t \text{ appears in document } d}{\text{Total number of terms in document } d}$$

**2. Frecuencia inversa de documentos (IDF):** Reduce la importancia de las palabras comunes en varios documentos, a la vez que aumenta la de las palabras poco frecuentes. Si un término aparece en menos documentos, es más probable que sea significativo y específico.
$$IDF(t, D) = \log \frac{\text{Total number of documents in corpus } D}{\text{Number of documents containing term } t}$$

Este equilibrio permite a TF-IDF resaltar los términos que son frecuentes dentro de un documento específico y que, a la vez, son distintivos en todo el texto, lo que la convierte en una herramienta útil para tareas como la clasificación en buscadores, la clasificación de textos y la extracción de palabras clave.

### Paso 1: Calcular la matriz TF-IDF para el corpus `data/01_corpus_turismo_500.txt`

```python
from pathlib import Path

# Cargamos el corpus: 01_corpus_turismo_500.txt
path_corpus_turismo = Path("data/01_corpus_turismo_500.txt")

# Trabajeromos como un documento independediente cada linea
with open(path_corpus_turismo, "r", encoding="utf-8") as f:
    corpus = f.readlines()

len(corpus)
```

En este paso convertimos cada línea del corpus en una representación numérica TF-IDF.

- `TfidfVectorizer()`: construye el vocabulario y calcula el peso TF-IDF de cada término.
- `fit_transform(corpus)`: aprende el vocabulario a partir del corpus y genera la matriz dispersa.
- `get_feature_names_out()`: devuelve los términos que quedaron en el vocabulario final.
- `pd.DataFrame(...)`: convierte la matriz a una tabla para inspeccionarla con facilidad.

El resultado permite analizar qué palabras son más relevantes dentro de cada documento y en todo el corpus.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Inicializamos el vectorizador TF-IDF
# Nota: Se puede usar TfidfVectorizer(stop_words='spanish') para eliminar palabras vacías
vectorizer= TfidfVectorizer()

# Ajustamos el vectorizador y transformamos el corpus en una matriz TF-IDF
# matriz_r es una matriz dispersa (sparse) de dimensiones (documentos, términos)
matriz_r = vectorizer.fit_transform(corpus)

# Obtenemos los nombres de todos los términos (palabras) del vocabulario
terminos = vectorizer.get_feature_names_out()

df = pd.DataFrame(matriz_r.toarray(), columns=terminos)
print("Tamaño de la matriz: ", matriz_r.shape)
print("Matriz TF-IDF:\n")
df
```

```python
# Calculamos el promedio de TF-IDF de cada término a través de todos los documentos
# Ordenamos de mayor a menor y mostramos los 10 términos más importantes globalmente
df.mean(axis=0).sort_values(ascending=False).head(10)
```

```python
# Obtenemos el primer documento y mostramos sus 10 términos con mayor valor TF-IDF
# Esto indica qué palabras son más distintivas para ese documento específico
doc1 = df.iloc[0].sort_values(ascending=False).head(10)
doc1
```

### Paso 2: Construir el corpus `Gutenberg 1000`

El corpus `Gutenberg 1000` es un corpus compuesto por 1000 libros de Gutenberg Project

```python
import os
from pathlib import Path

# Cargamos el corpus: books1000, que es un corpus compuesto por 1000 libros de Gutenberg Project
path_corpus_gutenberg1000 = Path("data/gutenberg/books1000")
books1000 = os.listdir(path_corpus_gutenberg1000)

corpusbooks1000 = []

for archivo in books1000:
    path_archivo = path_corpus_gutenberg1000/archivo
    with open(path_archivo, "r", encoding="utf-8") as f:
        contenido = f.read()
        corpusbooks1000.append(contenido)

print(f"Número de libros en el corpus: {len(corpusbooks1000)}")
print(f"Primeros 500 caracteres del libro 1:\n{corpusbooks1000[0][:500]}")
```

### Paso 3: Calcular la matriz TF-IDF para el corpus `Gutenberg 1000`

Este bloque construye el modelo TF-IDF sobre el corpus `corpusbooks1000`.

- `TfidfVectorizer(stop_words='english', max_df=0.5, min_df=2)`: elimina palabras vacías en inglés y filtra términos demasiado frecuentes o demasiado raros.
- `fit_transform(corpusbooks1000)`: aprende el vocabulario del corpus y genera la matriz dispersa TF-IDF.
- `get_feature_names_out()`: recupera los términos que forman el vocabulario final.
- `matriz_r.shape`: muestra cuántos documentos y términos quedaron en la representación final.

El objetivo es obtener una representación numérica del corpus para poder comparar documentos y analizar los términos más relevantes.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# Inicializamos el vectorizador TF-IDF
vectorizer= TfidfVectorizer(stop_words='english',max_df=0.7, min_df=2)

matriz_r = vectorizer.fit_transform(corpusbooks1000)

# Obtenemos los nombres de todos los términos (palabras) del vocabulario
terminos = vectorizer.get_feature_names_out()

print("Tamaño de la matriz: ", matriz_r.shape)
print("Número de términos únicos: ", len(terminos))
```

### Paso 4: Programar una función `buscar()` para el corpus `Gutenberg 1000`

```python
from sklearn.metrics.pairwise import cosine_similarity

def buscar(query, vectorizador, matriz, nombres, top_k=5):
    """Busca los libros más similares a una consulta.

    Args:
        query: Texto a buscar.
        vectorizador: `TfidfVectorizer` ya entrenado.
        matriz: Matriz TF-IDF del corpus.
        nombres: Lista de nombres de libros.
        top_k: Número de resultados a devolver.

    Returns:
        Lista de libros más similares.
    """

    # Convertir query a vector TF-IDF
    query_vec = vectorizador.transform([query])

    # Calcular similitud coseno
    similitudes = cosine_similarity(query_vec, matriz).flatten()

    # Obtener índices ordenados
    top_indices = similitudes.argsort()[::-1][:top_k]

    print(f'Resultados para: "{query}"\n')

    for i in top_indices:
        print(f'{nombres[i]:<50} similitud = {similitudes[i]:.4f}')
```

```python
buscar("warfare", vectorizer, matriz_r, books1000)
```
